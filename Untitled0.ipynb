```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
data['diagnosis']=lb.fit_transform(data['diagnosis'])
```

```python
x=data.iloc[:,2:-1]
y=data.diagnosis
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
scaled_x=sc.fit_transform(x)
x=pd.DataFrame(scaled_x,columns=x.columns)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.neural_network import MLPClassifier
```

```python
from sklearn.neural_network import MLPClassifier
model=MLPClassifier(hidden_layer_sizes=(10,4),
                    learning_rate_init=0.04,solver='adam',
                    max_iter=200,
                    random_state=2)
model.fit(x_train,y_train)
y_predict=model.predict(x_test)
y_train_predict=model.predict(x_train)
```

```python
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras import Input
from tensorflow.keras.layers import Dense
from keras.callbacks import ModelCheckpoint
```

```python

model = Sequential()
model.add(Input(shape=(30,)))
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='rmsprop',metrics=['accuracy'])
hist=model.fit(x_train,y_train,batch_size=32,epochs=15,validation_data=(x_test,y_test),verbose=1,shuffle=True)
```

```python
x_train.shape
```

```python
# prompt: accuracy

model.evaluate(x_test,y_test)[1]
```

```python
# prompt: validation

from sklearn.metrics import confusion_matrix, classification_report

y_pred = model.predict(x_test)

# Convert the predictions to binary values
y_pred_binary = [int(round(p[0])) for p in y_pred]

# Print the confusion matrix
print(confusion_matrix(y_test, y_pred_binary))

# Print the classification report
print(classification_report(y_test, y_pred_binary))
```

```python
# prompt: when to use recall and precision

Recall is a measure of how well a model identifies positive cases. It is calculated as the number of true positives divided by the sum of true positives and false negatives. Precision is a measure of how well a model avoids identifying negative cases as positive. It is calculated as the number of true positives divided by the sum of true positives and false positives.

Recall is more important than precision when the cost of missing a positive case is high. For example, in a medical diagnosis scenario, it is more important to correctly identify all patients with a disease, even if it means that some healthy patients are incorrectly identified as having the disease.

Precision is more important than recall when the cost of incorrectly identifying a negative case as positive is high. For example, in a fraud detection scenario, it is more important to avoid incorrectly identifying innocent people as fraudsters, even if it means that some fraudsters are not caught.

In general, the choice of which metric to use depends on the specific context and the relative costs of false positives and false negatives.
```
